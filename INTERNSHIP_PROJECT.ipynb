# Sales Data Analysis using Python

## Introduction
This project analyzes sales data to understand product performance, outlet performance, and factors affecting sales.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd

data = pd.read_csv('Big Sales Data.csv')
data.head()
```

## Data Understanding
This section explores the dataset structure, data types, and missing values.


```python
data.info()
```

```python
data.describe()
```

This provides statistical summary of numerical columns such as mean, minimum, maximum, and standard deviation.


```python
data.isnull().sum()
```

## Data Cleaning
Missing values in Item_Weight are filled using the mean value to maintain data consistency.


```python
data['Item_Weight'] = data['Item_Weight'].fillna(data['Item_Weight'].mean())

```

```python
data.isnull().sum()
```

## Data Analysis


```python
top_items = data.groupby('Item_Type')['Item_Outlet_Sales'].sum().sort_values(ascending=False).head(5)
print(top_items)
```

The analysis shows that Snack Foods and Fruits & Vegetables are the top-selling item categories, indicating high customer demand for these products.


```python
outlet_sales = data.groupby('Outlet_Type')['Item_Outlet_Sales'].sum()
print(outlet_sales)
```

Supermarket Type1 generates the highest sales among all outlet types.


```python
location_sales = data.groupby('Outlet_Location_Type')['Item_Outlet_Sales'].sum()
print(location_sales)
```

Tier 3 locations contribute the highest sales compared to other regions.


```python
top_items.plot(kind='bar')
plt.title('Top 5 Item Types by Sales')
plt.xlabel('Item Type')
plt.ylabel('Sales')
plt.show()
```

```python
outlet_sales.plot(kind='bar')
plt.title('Sales by Outlet Type')
plt.xlabel('Outlet Type')
plt.ylabel('Sales')
plt.show()
```

```python
location_sales.plot(kind='bar')
plt.title('Sales by Location Type')
plt.xlabel('Location')
plt.ylabel('Sales')
plt.show()
```

## Results

- Top item types contribute significantly to total sales.
- Supermarket Type outlets generate higher sales compared to grocery stores.
- Tier 3 locations show higher sales performance.


## Conclusion
